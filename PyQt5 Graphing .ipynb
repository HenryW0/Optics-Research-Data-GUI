```python
from PyQt5 import QtWidgets, QtCore, QtGui
from pyqtgraph import PlotWidget, plot
from PyQt5.QtWidgets import QPushButton, QWidget
import pyqtgraph as pg
import sys  # We need sys so that we can pass argv to QApplication
import os
import time


'''Code for Class is based on https://www.pythonguis.com/tutorials/plotting-pyqtgraph/ however changes were made so the 
inputs are from the serial port instead so we can read multiple voltages'''

import serial
import numpy as np

time_start = time.time()

ser = serial.Serial('COM6',  9600)

class MainWindow(QtWidgets.QMainWindow):
    def __init__(self, *args, **kwargs):
        super(MainWindow, self).__init__(*args, **kwargs)
        self.graphWidget = pg.PlotWidget()
        self.setCentralWidget(self.graphWidget)
        self.setWindowTitle('Lissajous Figure')
        
        self.a = []
        self.b = []
        self.CC = 0
        self.count = 0
        
        self.graphWidget.setBackground((0,0,0))
        
        self.data_line = self.graphWidget.plot(self.a, self.b, pen=None, symbol = 'o', symbolBrush = (255, 255, 255))
        self.data_line2 = self.graphWidget.plot(self.a[self.count:], self.b[self.count:], pen=None, symbol = 'o', symbolBrush = (0, 255, 0))
        
        
        self.saveBtn = QtGui.QPushButton("Save")
        self.saveBtn.setParent(self.graphWidget)
        self.saveBtn.move(50, 0)
        self.saveBtn.clicked.connect(self.Save)
        self.saveBtn.show()
        
        self.timer = QtCore.QTimer()
        self.timer.setInterval(0)
        self.timer.timeout.connect(self.update_plot_data)
        self.timer.start()
        
        
    def Save(self):
        print("Saved")
        np.save("PD_A", self.a)
        np.save("PD_B", self.b)
        np.save("Time", [-1])
        print(time.time() - time_start)
        
    def update_plot_data(self):
        
        line = ser.readline()
        line_list = line.split()
        map_object = map(float, line_list) #converts the values from bytes into floats
        voltages = list(map_object) #converts map object into a list
        
        self.b.append(voltages[0])  # Add a new serial value.
        self.a.append(voltages[1])
        
        self.data_line.setData(self.a[:self.count], self.b[:self.count], symbolBrush = pg.intColor(0), symbolPen = (255, 0, 0))  # Update the data
        self.data_line2.setData(self.a[self.count:], self.b[self.count:], symbolBrush = (0, 255, 0), symbolPen = (0, 255, 0))
        self.count += 1
         
        
        
app = QtWidgets.QApplication(sys.argv)
w = MainWindow()
w.show()
sys.exit(app.exec_())
```

```python
import numpy as np
import matplotlib.pyplot as plt
PD_A = np.load("PD_A.npy")
PD_B = np.load("PD_B.npy")
speed = 0.25


try:
    t = np.load("Time.npy")[0]
    
    if t < 0:
        z = np.arange(0, len(PD_A) * speed, speed)
        print("L plot used")
    else :
        z = np.arange(0, (t) * speed, speed)
        PD_A = PD_A[: t]
        PD_B = PD_B[: t]
        print("T plot used")
        
except:
    print("This should not get printined")

plt.style.use('ggplot')

fig1 = plt.figure(figsize = (10, 10))
plt.plot(z, PD_A, color = 'red', label = "PD_A")
plt.plot(z, PD_B, color = 'blue', label = "PD_B")
plt.legend()
plt.title("Voltages vs Time")
plt.xlabel("Time (s)")
plt.ylabel("Voltage (V)")
#plt.savefig("VoltageTimeHighContrast")
plt.show()

#plt.style.use('default')



fig2 = plt.figure(figsize = (10, 10))
points = plt.scatter(PD_A, PD_B, c = z, cmap = "jet")
plt.rcParams['axes.grid'] = False #Set to  false first to avoid error with color bar
cbar = plt.colorbar(points)
plt.rcParams['axes.grid'] = True
cbar.set_label("Time (s)")
cbar.minorticks_on()
#cbar.set_ticks([0, 20, 40, 60, 80, 100])
#cbar.set_ticklabels(["0", "20", "40", "60", "80", "100"])
plt.title("Photodetector B vs Photodetector A")
plt.xlabel("PD_A Voltage (V)")
plt.ylabel("PD_B Voltage (V)")
#plt.savefig("LissajousHighContrast")
plt.show()
```
